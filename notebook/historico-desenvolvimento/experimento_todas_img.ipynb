```python
import time
import sys, os
import numpy as np
from PIL import Image

sys.path.insert(0, os.path.abspath(".."))
from src import config

for classe in ["glioma", "meningioma", "notumor", "pituitary"]:
    pasta = os.path.join(config.TRAIN_DIR, classe)
    arquivos = sorted(os.listdir(pasta))
    inicio = time.time()
    for fname in arquivos:
        fpath = os.path.join(pasta, fname)
        with Image.open(fpath) as img:
            img = img.convert("L").resize((config.IMG_SIZE, config.IMG_SIZE))
            vetor = np.asarray(img, dtype=np.float32).flatten() / 255.0
    fim = time.time()
    print(f"{classe}: {fim-inicio:.2f}s para {len(arquivos)} imagens ({(fim-inicio)/len(arquivos)*1000:.1f}ms/imagem)")
```
