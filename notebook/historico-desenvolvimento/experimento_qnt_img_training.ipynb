```python
import sys, os
sys.path.insert(0, os.path.abspath(".."))
from src import config

for classe in ["glioma", "meningioma", "notumor", "pituitary"]:
    pasta = os.path.join(config.TRAIN_DIR, classe)
    todos = os.listdir(pasta)
    imagens = [f for f in todos if f.lower().endswith((".jpg", ".jpeg", ".png"))]
    nao_imagens = [f for f in todos if f not in imagens]
    print(f"{classe}: {len(todos)} arquivos | {len(imagens)} imagens | nao-imagem: {nao_imagens}")
```
